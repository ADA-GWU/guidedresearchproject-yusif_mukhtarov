```python
img = cv2.imread('../data/test/0026.jpg')

rotation = [30, 60, 90, 120, 150, 180]

# Create a figure with a 2x3 grid of subplots
fig, axes = plt.subplots(2, 3, dpi=300)

# Iterate through each rotation degree and plot the rotated image in the corresponding subplot
for i, r in enumerate(rotation):
    # Perform rotation
    rows, cols = img.shape[:2]
    M = cv2.getRotationMatrix2D((cols / 2, rows / 2), r, 1)
    img_rotated = cv2.warpAffine(img, M, (cols, rows))
    print(img_rotated.shape)
    
    # Determine the subplot indices
    row_idx = i // 3
    col_idx = i % 3
    
    # Turn off axes in the subplot
    axes[row_idx, col_idx].axis("off")
    
    # Display the rotated image in the corresponding subplot
    img_rotated_final = cv2.cvtColor(img_rotated, cv2.COLOR_BGR2RGB)
    print(img_rotated_final.shape)
    axes[row_idx, col_idx].imshow(img_rotated_final)
    axes[row_idx, col_idx].set_title(f"Rotation: {r}")

# Adjust the spacing between subplots
plt.subplots_adjust(hspace=0.4, wspace=0.2)

# Show the figure with all the rotated images
plt.show()
```
